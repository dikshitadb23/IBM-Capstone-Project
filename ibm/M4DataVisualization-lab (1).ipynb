# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite")
```

```python
query = """select(ConvertedComp) from master"""
ConvertedComp = pd.read_sql_query(query,conn)
ConvertedComp.plot(kind = 'hist')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
query1 = """SELECT COUNT(Age) FROM master"""
Age= pd.read_sql_query(query1, conn)
Age.plot(kind='box', figsize=(10, 7), notch = True, widths = 5.0 )
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
query2 = """SELECT COUNT(WorkWeekHrs) FROM master"""
WorkWeekHrs = pd.read_sql_query(query2, conn)
# sns.scatterplot( x = Age, y = WorkWeekHrs)
plt.scatter( x = WorkWeekHrs, y = Age,   alpha=0.5)
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

query1 = """SELECT COUNT(Age) FROM master"""
Age= pd.read_sql_query(query1, conn)
query2 = """SELECT COUNT(WorkWeekHrs) FROM master"""
WorkWeekHrs = pd.read_sql_query(query2, conn)
query3 = """SELECT COUNT(CodeRevHrs) FROM master"""
CodeRevHrs = pd.read_sql_query(query3, conn)

# sns.scatterplot( x = Age, y = WorkWeekHrs)
plt.scatter( x = WorkWeekHrs, y = CodeRevHrs,  s = Age )
# plt.plot(kind = 'scatter' , x = WorkWeekHrs, y = CodeRevHrs, s = Age, )
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query4="""
SELECT DatabaseDesireNextYear, count(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc limit 5
"""

DatabaseDesireNextYear= pd.read_sql_query(query4, conn)
x = DatabaseDesireNextYear['count']
label = DatabaseDesireNextYear['DatabaseDesireNextYear']
plt.pie(x, labels=label, autopct='%1.1f%%',radius=1.5)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
query5 = """
SELECT Avg(WorkWeekHrs), Avg(CodeRevHrs), Age 
FROM master 
where Age between 30 and 35
group by Age
"""

z = pd.read_sql_query(query5, conn)
WorkWeekHrs= z['Avg(WorkWeekHrs)']
CodeRevHrs= z['Avg(CodeRevHrs)']
Age= z['Age']

fig, ax = plt.subplots()

ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
query7 = """SELECT MainBranch, count(*)
FROM master
group by MainBranch"""
MainBranch = pd.read_sql_query(query7,conn)
MainBranch.plot(kind='barh')
plt.show()
```
